**Extracting All the tables From PDF at a time**

```python
doc= cam.read_pdf('/content/NIT_164.pdf', pages='all')
print('total table extracted:',doc.n)
```

```python
doc_df=[]
for i in range(len(doc)):
  doc_df.append(doc[i].df)
  doc_df[i].rename(columns=doc_df[i].iloc[0], inplace=True)
  doc_df[i].drop(doc_df[i].index[0], inplace=True)
  display(doc_df[i])

  
```
